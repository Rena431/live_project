task 2


```python
df.columns = [
    "employee_id",
    "role",
    "years_exp",
    "current_ctc",
    "previous_ctc",
    "company",
    "company_type",
    "skills",
    "location",
    "education_tier",
    "joining_year",
    "work_mode"
]

df.head()
```

```python
print(df["role"].unique())
```

```python
role_map = {
    "Backend Engineer":"SDE Backend",
    "Backend Dev":"SDE Backend",
    "BE":"SDE Backend",
    "Frontend Engineer":"SDE Frontend",
    "Frontend Dev":"SDE Frontend",
    "FE":"SDE Frontend",
    "Full Stack Engineer":"SDE Full-Stack",
    "Full Stack Dev":"SDE Full-Stack",
    "FS":"SDE Full-Stack",
    "DevOps":"DevOps Engineer",
    "UI Designer":"UI/UX Designer",
    "UX Designer":"UI/UX Designer"
}

df["role"] = df["role"].replace(role_map)
```

```python
df["company_type"] = df["company_type"].str.strip().str.title()
```

```python
tier_map = {
    "T1":"Tier 1",
    "Tier-1":"Tier 1",
    "1":"Tier 1",
    "T2":"Tier 2",
    "Tier-2":"Tier 2",
    "2":"Tier 2",
    "T3":"Tier 3",
    "Tier-3":"Tier 3",
    "3":"Tier 3"
}

df["education_tier"] = df["education_tier"].replace(tier_map)
```

```python
def parse_ctc(x):
    if pd.isna(x):
        return np.nan

    x = str(x).replace("₹", "")
    x = x.replace("LPA", "")
    x = x.replace(",", "")
    x = x.strip()

    value = float(x)

    if value > 1000:
        value = value / 100000

    return value
```

```python
df["current_ctc"] = df["current_ctc"].apply(parse_ctc)
df["previous_ctc"] = df["previous_ctc"].apply(parse_ctc)
```

```python
df = df.drop_duplicates()
print(df.dtypes)
print(df.dtypes)
print(df["role"].value_counts())
print(df["company_type"].value_counts())
print(df["education_tier"].value_counts())
print(df["work_mode"].value_counts())
```

task 3


```python
role_ctc = df.groupby("role")["current_ctc"].agg(["median","mean","min","max"])
role_ctc = role_ctc.sort_values(by="median", ascending=False)
print(role_ctc)
```

```python
backend = df[df["role"]=="SDE Backend"].copy()

bins = [-1,1,3,5,100]
labels = ["0-1","2-3","4-5","6+"]

backend["exp_band"] = pd.cut(backend["years_exp"], bins=bins, labels=labels)

exp_ctc = backend.groupby("exp_band")["current_ctc"].median()

print(exp_ctc)
```

```python
sde = df[df["role"].isin(["SDE Backend","SDE Frontend","SDE Full-Stack"])]

skills = ["AWS","ML","System Design","Kubernetes"]

for skill in skills:
    with_skill = sde[sde["skills"].str.contains(skill, case=False, na=False)]["current_ctc"].median()
    without_skill = sde[~sde["skills"].str.contains(skill, case=False, na=False)]["current_ctc"].median()

    premium=((with_skill-without_skill)/without_skill)*100

    print(skill)
    print("With Skill :",with_skill)
    print("Without Skill :",without_skill)
    print("Premium :",round(premium,2),"%")
    print()
```

```python
backend = df[df["role"]=="SDE Backend"]

company_ctc = backend.groupby("company_type")["current_ctc"].median()

print(company_ctc)

unicorn = company_ctc["Unicorn"]
mnc = company_ctc["Mnc"]

premium=((unicorn-mnc)/mnc)*100

print("Unicorn Premium :",round(premium,2),"%")
```

```python
bins=[-1,1,3,5,100]
labels=["0-1","2-3","4-5","6+"]

df["exp_band"]=pd.cut(df["years_exp"],bins=bins,labels=labels)

group_size=df.groupby(["role","company_type","exp_band"])["current_ctc"].transform("count")

group_median=df.groupby(["role","company_type","exp_band"])["current_ctc"].transform("median")

temp=df[group_size>=10].copy()

temp["group_median"]=group_median[group_size>=10]

temp["gap"]=temp["current_ctc"]-temp["group_median"]

underpaid=temp.sort_values("gap").head(10)

print(underpaid[["employee_id","role","company_type","years_exp","current_ctc","group_median","gap"]])
```
